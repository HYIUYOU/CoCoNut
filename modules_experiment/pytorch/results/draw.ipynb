```python
import pandas as pd
import matplotlib.pyplot as plt

def draw(file_path, rowindex, ax):
    # 读取 CSV 文件
    df = pd.read_csv(file_path)

    # 获取指定行的数据
    row = df.iloc[rowindex]

    # 获取 mean, max, 以及数据点的部分
    mean_value = row['mean']
    max_value = row['max']
    data_points = row[3:]  # 假设数据点从索引 3 开始

    # 绘制数据点的折线
    ax.plot(data_points, label='Data Points', marker='o', linestyle='-')

    # 绘制 mean 的水平线
    ax.axhline(y=mean_value, color='r', linestyle='--', label='Mean')

    # 绘制 max 的水平线
    ax.axhline(y=max_value, color='g', linestyle='-.', label='Max')

    # 找到最大值的索引
    max_index = data_points.idxmax()
    max_index = int(max_index)

    # 在图中标注最大值的位置
    ax.annotate(f'MAX: {round(data_points.iloc[max_index], 2)}, x={max_index}', 
                xy=(max_index, data_points.iloc[max_index]), 
                xytext=(max_index + 10, data_points.iloc[max_index] - 0.15),  # 设置偏移量，避免文字和点重叠
                arrowprops=dict(facecolor='red', edgecolor='red', 
                                arrowstyle='-|>,head_length=0.4,head_width=0.2',
                                shrinkB=0),
                fontsize=10,
                color='red'
                )

    # 设置横坐标的间隔，比如每隔20个点显示一个刻度标签
    ax.set_xticks(ticks=range(0, len(data_points), 20))

    # 设置标题和标签
    ax.set_xlabel('Forwards')
    ax.set_ylabel('Latency (s)')
    #ax.set_title(f'Module #{rowindex}')
    ax.legend()

# 文件路径
file_paths = [
    '1_module_persist1_bs10_tokens128.csv',
    '2_module_persist1_bs10_tokens128.csv',
    '3_module_persist1_bs10_tokens128.csv',
    '4_module_persist1_bs10_tokens128.csv',
    '5_module_persist1_bs10_tokens128.csv',
    '6_module_persist1_bs10_tokens128.csv'
]

# 创建一个包含 3 行 2 列子图的画布
fig, axs = plt.subplots(3, 2, figsize=(15, 6))

# 使用循环绘制每个文件的数据到对应的子图中
for i, file_path in enumerate(file_paths):
    row, col = divmod(i, 2)
    draw(file_path, 0, axs[row, col])

# 调整布局以避免子图重叠
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.suptitle('Module #0')
# 显示所有图
plt.show()


```
