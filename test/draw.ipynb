```python
import matplotlib.pyplot as plt

def compare(*arrays, labels=None,title=''):
    """
    绘制多个折线图，每个折线图的标题是数组的名称。

    参数:
    *arrays: 传入多个数组，这些数组会被绘制成折线图。
    labels: 可选参数，传入一个列表，包含每个数组的标签（名称）。
    """
    # 创建一个图形
    plt.figure(figsize=(10, 6))

    # 遍历每个数组，绘制对应的折线
    for i, array in enumerate(arrays):
        label = labels[i] if labels else f"Array {i+1}"  # 使用传入的标签，默认是"Array i"
        plt.plot(array, label=label, marker='o', linestyle='-', alpha=0.7)
 # 绘制折线并使用标签作为图例

    # 添加标题
    plt.title(title)

    # 添加坐标轴标签
    plt.xlabel('Module Index')
    plt.ylabel('Latency (s)')

    # 显示图例
    plt.legend()

    # 显示图形
    plt.show()
```
